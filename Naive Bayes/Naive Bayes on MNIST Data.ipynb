## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.model_selection import cross_val_score
from sklearn.naive_bayes import GaussianNB,MultinomialNB
```

## Dataset Preparation

```python
digits = load_digits() #sklearn MNIST dataset
```

```python
X = digits.data
Y = digits.target
```

```python
print(X.shape)
print(Y.shape)
```

```python
plt.imshow(X[0].reshape((8,8)),cmap='gray')
plt.show()
print(Y[0])
```

```python
print(X[0])
```

## Train the models

```python
mnb = MultinomialNB() # for dataset with discrete features
gnb = GaussianNB() # for dataset with continuous features
```

```python
mnb.fit(X,Y)
gnb.fit(X,Y)
```

## Results

```python
print(mnb.score(X,Y))
print(gnb.score(X,Y))
```

### Since each feature in dataset is discrete (values from 0 to 15), hence MNB gives better results than GNB

```python
# 10 fold cross validation
print(cross_val_score(mnb,X,Y,scoring='accuracy',cv = 10).mean())
print(cross_val_score(gnb,X,Y,scoring='accuracy',cv = 10).mean())
```
